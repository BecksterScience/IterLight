## Clustering only based on Mission stated/pillars

```python
import pandas as pd

# Select and clean the text column
text_col = 'Mission / Stated Pillars'
missions = df[text_col].dropna().astype(str).reset_index(drop=True)

print(f"Number of mission statements: {len(missions)}")
missions.head()
```

```python
!pip install sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer

# Load a compact but strong model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Encode missions into embeddings (vectors)
embeddings = model.encode(missions, show_progress_bar=True)
embeddings.shape
```

```python
from sklearn.cluster import KMeans

k = 5  # choose roughly based on intuition or elbow plot later
kmeans = KMeans(n_clusters=k, random_state=42)
labels = kmeans.fit_predict(embeddings)

df_clusters = pd.DataFrame({'Mission': missions, 'Cluster': labels})
df_clusters.head()
```

```python
#!pip install hdbscan
import hdbscan

clusterer = hdbscan.HDBSCAN(min_cluster_size=7)
labels = clusterer.fit_predict(embeddings)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

inertias = []
K = range(2, 15)  # test between 2 and 15 clusters
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(embeddings)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(K, inertias, 'o-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2, random_state=42)
embedding_2d = pca.fit_transform(embeddings)

plt.figure(figsize=(10,7))
plt.scatter(embedding_2d[:,0], embedding_2d[:,1], c=labels, cmap='tab10', s=50)
plt.title('Mission Statement Clusters (PCA)')
plt.show()
```
